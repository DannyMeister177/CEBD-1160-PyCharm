```python
import pandas as pd
import matplotlib.pyplot as plt
import os

plt.style.use('dark_background')

df = pd.read_csv('data/boston/housing.data', sep='\s+')
df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 
              'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

fig, axes = plt.subplots(1, 1, figsize=(15,15))

axes.scatter(df['B'], df['LSTAT'], s=df['MEDV']**2, color='g', marker='o', alpha= 0.25)
axes.set_xlabel('Measure of Black population')
axes.set_ylabel('% lower status of the population')

axes.scatter(df['B'], df['PTRATIO'], s=df['MEDV']**1.3, color='b', marker='o', alpha= 0.2)
axes.set_xlabel('pupil-teacher ratio by town')
```
