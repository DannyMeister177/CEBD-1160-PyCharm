```python
import pandas as pd
import matplotlib.pyplot as plt
import os

plt.style.use('dark_background')

df = pd.read_csv('data/boston/housing.data', sep='\s+')
df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 
              'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

def normalize(col):
    return ((col - col.mean()) / col.std())*15

B_n = normalize(df['B'])
LSTAT_n = normalize(df['LSTAT'])
PTRATIO_n = normalize(df['PTRATIO'])
MEDV_n = normalize(df['MEDV'])

fig, axes = plt.subplots(1, 1, figsize=(15,15))

axes.scatter(B_n, LSTAT_n, label = 'Measure of Black population', s=MEDV_n**2, 
             color='g', marker='o', alpha= 0.3)
axes.set_xlabel('Measure of Black population')
axes.set_ylabel('')
axes.legend()

axes.scatter(B_n, PTRATIO_n, label = '% lower status of the population', s=MEDV_n**1.3, 
             color='b', marker='o', alpha= 0.3)
axes.legend()
```
